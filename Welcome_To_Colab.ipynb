```
d = {'a' : 1, 'b' : 2, 'c' : 3}
series1 = pd.Series(d)
print(series1)
series2 = pd.Series(d, index = ['a', 'b'])
print(series2)
series3 = pd.Series(d,index = ['a','x'],name = 'data series')
print(series3)
print('printing using the dictionary key: ',series1['b'])
```

```
l = [1,2,3]
ser1 = pd.Series(l)
print(ser1)
ser2 = pd.Series(l,index = ['x','y','z'],name = 'list data series')
print(ser2)
```

```
l1 = [1,2]
ser = pd.Series(l1, copy = False)
print('series before update: ')
print(ser)
print('-------------')
ser[0] = 20
print('original list',l1)
print('-------------')
print('series after update: ')
print(ser)
print('in case of python objects, the original data will not be changed independent of the copy parameter')
```

```
import numpy as np
l2 = np.array([1,2])
print('original list',l1)
print('-------------')
ser3 = pd.Series(l2, copy = False)
print('series before update: ')
print(ser3)
print('-------------')
ser3[0] = 20
print('series after update: ')
print(ser)
print('in case of numpy arrays, the original data will  be changed if the copy parameter is set to False or the origibal data will be same in the numpy array')
```

```
#dataframes using dictionaries
dict = {'numbers' : [1,2,3], 'letters' : ['a','b','c']}
dic_dif = pd.DataFrame(dict)
print(dic_dif)
print('------------------------')
dic_dif = pd.DataFrame.from_dict(dict)
print(dic_dif)
print('------------------------')
dic_df = pd.DataFrame.from_dict(dict, orient = 'index')
print(dic_df)
print('------------------------')
#if only orient parameter is mentioned the key value is the index and the column titles are set to default
dic_df = pd.DataFrame.from_dict(dict, orient = 'index', columns = ['A','B','C'])
print(dic_df)

print('this orient parameter works only with the from_dict function')
```

```
url = 'https://raw.githubusercontent.com/pandas-dev/pandas/main/pandas/tests/io/data/csv/tips.csv'
csv_df = pd.read_csv(url, index_col = ['total_bill'])
print('before reset')
print(csv_df)
print('-------------------------------------------------------------')
# to reset the index to default we can either re-assign it to the variable or can use the inplace parameter in reset_index function
csv_df.reset_index(inplace = True)
print('after reset')
print(csv_df)
```
